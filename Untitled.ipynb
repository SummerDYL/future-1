```python
from pyecharts import Scatter
scatter = Scatter("价格——行为")
scatter.add("", x_axis=df.p1, y_axis=df.p2, xaxis_name="P(看涨|涨价)", yaxis_name="P(看跌|跌价)", 
            xaxis_name_pos="middle", yaxis_name_pos="end", yaxis_name_gap=18)
scatter
```

```python
# 专看投机人群
momentum = Scatter("动量型")
df1 = df[(df.p1>0.5) & (df.p2>0.5)]
momentum.add("", x_axis=df1.p1, y_axis=df1.p2, xaxis_name="P(看涨|涨价)", yaxis_name="P(看跌|跌价)", 
            xaxis_name_pos="middle", yaxis_name_pos="end", yaxis_name_gap=18, xaxis_min=0.5, yaxis_min=0.5)
momentum
```
