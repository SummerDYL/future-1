```python
import MySQLdb
conn = MySQLdb.connect(host='219.224.169.45', user='fairy', passwd='fairy_tail', db='investor', charset='utf8')
cursor = conn.cursor()
sql = 'select profit-commission from profit_annual where year=%s and commission>0'    # 手续费大于0筛除了全年无交易的人
data = {}
for year in range(2014,2017):
    data[year] = []
    cursor.execute(sql, (year,))
    for row in cursor.fetchall():
        data[year].append(row[0])
```

```python
from pyecharts import Bar
bar = Bar("年收益分布", width=950)
bin_width = 10**4
left = (int(min(min(data[2014]),min(data[2015]),min(data[2016])) / bin_width)-1) * bin_width
bin_num = int((max(max(data[2014]), max(data[2015]), max(data[2016]))-left) / bin_width) + 1
X = ['{0}~{1}'.format(left+i*bin_width, left+(i+1)*bin_width) for i in range(bin_num)]
Y = {}
for year in range(2014,2017):
    Y[year] = [0] * bin_num
    for x in data[year]:
        Y[year][int((x-left)/bin_width)] += 1
    bar.add(str(year), X, Y[year], is_datazoom_show=True, xaxis_name='收益/元', xaxis_name_pos='end', yaxis_name='人数', yaxis_name_pos='end')
bar
```

```python
bar2 = Bar('年收益分布', width=950)
YY = {}
for year in range(2014,2017):
    tmp = sum(Y[year])
    YY[year] = [round(float(x)/tmp,3) for x in Y[year]]
    bar2.add(str(year), X, YY[year], is_datazoom_show=True, xaxis_name='收益/元', xaxis_name_pos='end', yaxis_name='人数(比例)', yaxis_name_pos='end')
bar2
```
