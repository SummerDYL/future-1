```python
import datetime
account = '488854682822'

import MySQLdb
conn = MySQLdb.connect(host='219.224.169.45', user='lizimeng', passwd='codegeass', db='investor', charset='utf8')
cursor = conn.cursor()
sql = 'select sum(total_times) from trade_times_daily where account=%s and year(date)=%s and month(date)=%s'
total_times = {}
for year in range(2014,2017):
    total_times[year] = {}
    for month in range(1,13):
        cursor.execute(sql, (account,year,month,))
        total_times[year][month] = int(cursor.fetchone()[0])
cursor.close()
conn.close()
```

```python
from pyecharts import HeatMap
x_axis = [str(i) for i in range(1,13)]
y_axis = [str(i) for i in range(2014,2017)]
data = [[i,j,total_times[j+2014][i+1]] for i in range(12) for j in range(3)]
heatmap = HeatMap()
heatmap.add("", x_axis, y_axis, data, is_visualmap=True, visual_range=[0,1000], visual_text_color="#000", 
            visual_orient='horizontal', visual_pos='center', label_pos='inside', tooltip_tragger='none')
heatmap
```
